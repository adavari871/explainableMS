```python
import nibabel as nib
import numpy as np
from scipy.ndimage.interpolation import zoom
```

```python
scan = nib.load("/analysis/ritter/misc/masks/Neuromorphometrics/tpm/labels_Neuromorphometrics.nii")
neuromorph_map = scan.get_data()
nmm_map = np.zeros(shape=(96, 114, 96))

for lbl_idx in np.unique(neuromorph_map):
    print(lbl_idx)
    nmm_map_lbl = neuromorph_map.copy()
    nmm_map_lbl[lbl_idx!=neuromorph_map] = 0
    nmm_map_lbl[lbl_idx==neuromorph_map] = 1
    zoomed_lbl = zoom(nmm_map_lbl, 0.7895, order=0)
    zoomed_lbl[zoomed_lbl!=1] = 0
    nmm_map += zoomed_lbl*lbl_idx
    print(np.unique(nmm_map))
```

```python
output_map = nib.Nifti1Image(nmm_map, scan.affine)
nib.save(output_map, "/analysis/fabiane/scaled_nmm_order0_again.nii")
```

```python
scan = nib.load("/analysis/fabiane/atlases/JHU/JHU-ICBM-labels-1mm.nii.gz")
atlas_map = scan.get_data()
atl_map = np.zeros(shape=(96, 114, 96))

for lbl_idx in np.unique(atlas_map):
    print(lbl_idx)
    atl_map_lbl = atlas_map.copy()
    atl_map_lbl[lbl_idx!=atlas_map] = 0
    atl_map_lbl[lbl_idx==atlas_map] = 1
    zoomed_lbl = zoom(atl_map_lbl, 0.525, order=0)
    zoomed_lbl[zoomed_lbl!=1] = 0
    atl_map += zoomed_lbl*lbl_idx
    print(np.unique(atl_map))
```

```python
output_map = nib.Nifti1Image(atl_map, scan.affine)
nib.save(output_map, "/analysis/fabiane/scaled_JHU_WM_order0.nii")
```
